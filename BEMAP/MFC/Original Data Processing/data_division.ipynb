```python
#归一化事件值
def nor(dt,len_app):
    nor_data = pd.DataFrame()
    event_list = dt.columns.tolist()
    for i in range(len(event_list)):
        if dt.max()[event_list[i]] == 0:
            nor_data[event_list[i]] = [0]*len_app
        else:
            nor_data[event_list[i]] = (dt[event_list[i]]-dt.min()[event_list[i]])/(dt.max()[event_list[i]]-dt.min()[event_list[i]])
    return nor_data
```

```python
#数据格式转化
def data_trans(nor_data,data_ipc,outputpath,action):
    list_i = []
    for i in range(100):
        for j in range(1,6):
            tx = 'ipc'+ str(i) + '_' +str(j)
            list_i.append(tx)

    data_all_initial = pd.read_csv('initial_dataframe_pred.csv', index_col=0)
    ipc_list = list_i

    for i in range(len(name_list)*5):
        for m in range(len_col):
            data_all_initial.loc[i*len_col+m, event_list[m]] = nor_data.loc[i, event_list[m]]
            data_all_initial.loc[i*len_col+m, 'IPC'] = data_ipc.loc[m, ipc_list[i]]
        #data_all_initial.loc[i*len_col:(i+1)*len_col, 'IPC'] = data_ipc.loc[0:len_col, ipc_list[i]]
    data_all_initial.to_csv(outputpath+'/'+outputpath+'_'+action+'_pred.csv',sep=',',index=True,header=True)
    return data_all_initial
```
